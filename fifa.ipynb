```python

player_data = data[data['Name'] == 'L. Messi'].iloc[0]
stats = ['Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 
         'Dribbling', 'Curve', 'FKAccuracy', 'LongPassing', 'BallControl']
values = player_data[stats].values.tolist()
values+=values[:1]
categories=stats
N=len(categories)

angles=[n/float(N)*2*3.14 for n in range(N)]

angles+=angles[:1]

ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)   
ax.plot(angles, values)
ax.fill(angles, values, 'teal', alpha=0.25)
plt.title(f"{player_data['Name']}'s Skill Set", size=20, color='blue', y=1.1)
plt.show()
```
